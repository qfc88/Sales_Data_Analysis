## Feature engineering

```python
from pyspark.sql.functions import datediff, date_format
```

```python
year = Window.partitionBy('YEAR_ID')
quarter = Window.partitionBy('YEAR_ID', 'QTR_ID')
month = Window.partitionBy('YEAR_ID', 'MONTH_ID')
product = Window.partitionBy('PRODUCTLINE')
product_month = Window.partitionBy('PRODUCTLINE', 'YEAR_ID', 'MONTH_ID')
product_quarter = Window.partitionBy('PRODUCTLINE', 'YEAR_ID', 'QTR_ID')
```

```python
df = df.withColumn('yearly_avg_quantity', avg('QUANTITYORDERED').over(year)) \
    .withColumn('quarter_avg_quantity', avg('QUANTITYORDERED').over(quarter)) \
    .withColumn('monthly_avg_quantity', avg('QUANTITYORDERED').over(month))\
    .withColumn('product_month_avg', avg('QUANTITYORDERED').over(product_month)) \
    .withColumn('product_quarter_avg', avg('QUANTITYORDERED').over(product_quarter)) \
    .withColumn('product_avg_quantity', avg('QUANTITYORDERED').over(product)) \
    .withColumn('product_total_orders', count('ORDERNUMBER').over(product))
```

```python
df = df.withColumn('price_to_msrp_ratio', col('PRICEEACH') / col('MSRP'))
```

```python
df = df.withColumn('discount_percentage', (1 - col('PRICEEACH') / col('MSRP')) * 100)
```

```python
df = df.withColumn('price_level', when(col('price_to_msrp_ratio') <= 0.6, 'Low')\
                   .when(col('price_to_msrp_ratio') <= 0.8, 'Medium')\
                   .otherwise('High'))
```

### Encoding

```python
from pyspark.ml.feature import VectorAssembler, StringIndexer, OneHotEncoder
from pyspark.ml import Pipeline
```

```python
train_data, test_data = df.randomSplit([0.8, 0.2], seed=42)
```

```python
categorical_features = ['PRODUCTLINE', 'DEALSIZE', 'price_level']
```

```python
indexers = [
        StringIndexer(
            inputCol=col, 
            outputCol=f"{col.lower()}_index",
            handleInvalid="keep"
        ) for col in categorical_features
    ]
```

```python
encoders = [
        OneHotEncoder(
            inputCol=f"{col.lower()}_index",
            outputCol=f"{col.lower()}_vec"
        ) for col in categorical_features
    ]
```

```python
df.printSchema()
```

```python
feature_cols = [
        'YEAR_ID', 'QTR_ID', 'MONTH_ID',
        'yearly_avg_quantity', 'quarter_avg_quantity', 'monthly_avg_quantity',
        'product_month_avg', 'product_quarter_avg',
        'product_avg_quantity', 'product_total_orders',
        'PRICEEACH', 'MSRP', 'price_to_msrp_ratio', 'discount_percentage',
        'productline_vec', 'dealsize_vec', 'price_level_vec'
    ]
```

```python
assembler = VectorAssembler(inputCols=feature_cols, outputCol='features')
```

## Data processing

```python
from pyspark.ml.regression import LinearRegression, RandomForestRegressor, GBTRegressor, DecisionTreeRegressor
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.clustering import KMeans
```

```python
lr = LinearRegression(featuresCol='features', labelCol='QUANTITYORDERED',
                     maxIter=10, regParam=0.3, elasticNetParam=0.8)
lr_pipe = Pipeline(stages=indexers + encoders + [assembler, lr])
lr_model = lr_pipe.fit(train_data)
```

```python
rf = RandomForestRegressor(featuresCol='features', labelCol='QUANTITYORDERED', numTrees=200, maxDepth=10, minInstancesPerNode=2, maxBins=64)
rf_pipe = Pipeline(stages=indexers + encoders + [assembler, rf])
rf_model = rf_pipe.fit(train_data)
```

```python
gbt = GBTRegressor(featuresCol='features', labelCol='QUANTITYORDERED', maxIter=50, maxDepth=8, stepSize=0.1)
gbt_pipe = Pipeline(stages=indexers + encoders + [assembler, gbt])
gbt_model = gbt_pipe.fit(train_data)
```

```python
dtg = DecisionTreeRegressor(featuresCol='features', labelCol='QUANTITYORDERED', maxDepth=10, maxBins=32)
dtg_pipe = Pipeline(stages=indexers + encoders + [assembler, dtg])
dtg_model = dtg_pipe.fit(train_data)
```

```python
evaluator = RegressionEvaluator(labelCol='QUANTITYORDERED', 
                               predictionCol='prediction', 
                               metricName='rmse')
```

```python
lr_predictions = lr_model.transform(test_data)
lr_rmse = evaluator.setMetricName('rmse').evaluate(lr_predictions)
lr_r2 = evaluator.setMetricName('r2').evaluate(lr_predictions)
```

```python
print("Linear Regression Results:")
print(f"RMSE: {lr_rmse:.2f}")
print(f"R2: {lr_r2:.2f}")
```

```python
rf_predictions = rf_model.transform(test_data)
rf_rmse = evaluator.setMetricName('rmse').evaluate(rf_predictions)
rf_r2 = evaluator.setMetricName('r2').evaluate(rf_predictions)
```

```python
print("Random Forest Results:")
print(f"RMSE: {rf_rmse:.2f}")
print(f"R2: {rf_r2:.2f}")
```

```python
gbt_predictions = gbt_model.transform(test_data)
gbt_rmse = evaluator.setMetricName('rmse').evaluate(gbt_predictions)
gbt_r2 = evaluator.setMetricName('r2').evaluate(gbt_predictions)
```

```python
print("Gradient Boosted Trees Results:")
print(f"RMSE: {gbt_rmse:.2f}")
print(f"R2: {gbt_r2:.2f}")
```

```python
dtg_predictions = dtg_model.transform(test_data)
dtg_rmse = evaluator.setMetricName('rmse').evaluate(dtg_predictions)
dtg_r2 = evaluator.setMetricName('r2').evaluate(dtg_predictions)
```

```python
print("Decision Trees Results:")
print(f"RMSE: {dtg_rmse:.2f}")
print(f"R2: {dtg_r2:.2f}")
```

```python
models = {
    'Linear Regression': (lr_model, lr_rmse),
    'Random Forest': (rf_model, rf_rmse),
    'GBT': (gbt_model, gbt_rmse),
    'DTG': (dtg_model, dtg_rmse)
}
```

```python
best_rmse = float('inf')  
best_model_name = None
best_model = None

for model_name, (model, rmse) in models.items():
    if rmse < best_rmse:
        best_rmse = rmse
        best_model_name = model_name
        best_model = model
import shutil

model_path = "models/demand_prediction_model"

if os.path.exists(model_path):
    shutil.rmtree(model_path)

print(f"Best Model: {best_model_name} with RMSE: {best_rmse:.2f}")
best_model.save(model_path)
```
